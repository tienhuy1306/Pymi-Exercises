# Exercise 11_4

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    data = np.random.randint(1, 1000000, 1000),
    columns = ['random'],
    index = pd.date_range(start = '2010/2/9', periods = 1000))
print(df)
```

```python
cum_sum = np.cumsum(df)
cum_sum.columns = ['cum_sum']
cum_sum['random'] = df['random']
print(cum_sum)
```

```python
cum_sum.plot()
```

```python
import numpy as np
import pandas as pd
acreage = pd.read_csv('/Users/admin/pyfml/exercises/dientich.csv')
acreage.columns = ['column']
data = acreage.column.str.split(';')[2:71].tolist()
acreage_new = pd.DataFrame(data = [i[1:6] for i in data],
                           columns = ['2011', '2012', '2013', '2014', '2015'],
                           index = [i[0] for i in data])
acreage_new = acreage_new.apply(pd.to_numeric) 
```
